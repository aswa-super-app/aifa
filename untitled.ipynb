### 🧠 Core Framework Logic Types

#### 1. Deterministic Logic (Rule-Based)
These logics do not use prompts in the traditional AI sense. Their "prompt" is a strict programmatic rule or mathematical equation. They are binary, 100% predictable, and fast.

*   **Regex / Pattern Matching Logic**
    *   **Core Rule:** `IF string matches <pattern>, RETURN match, ELSE RETURN None`
    *   **Use Case:** Extracting specific formats like an IC number (e.g., `\d{6}-\d{2}-\d{4}`).
*   **Validation Logic**
    *   **Core Rule:** `ASSERT type(X) == expected_type AND constraints_met(X)`
    *   **Use Case:** Checking if a parsed date actually exists on the calendar, or if an age is > 18.
*   **Calculation Logic**
    *   **Core Rule:** `IF A + B + C == Total, RETURN True, ELSE RETURN False`
    *   **Use Case:** Verifying that the Basic Contribution + Service Tax + Stamp Duty equals the Total Due on an invoice.
*   **Threshold / State Routing Logic**
    *   **Core Rule:** `IF ConfidenceScore > 0.85 THEN Route_A ELSE Route_B`
    *   **Use Case:** Deciding whether to auto-approve a document or send it to a human for manual review.

---

#### 2. AI Logic (Feedback / Probabilistic-Based)
These logics rely on Large Language Models (LLMs) or neural networks. Their behavior is guided by a **System Prompt** that defines the parameters of their decision-making.

*   **LLM Extraction Logic**
    *   **Core Prompt:** *"You are an expert data extractor. Extract the following fields: [List of Fields] from the provided text. Return ONLY a valid JSON object with these keys. If a field is missing, return 'null' for that key. Text: {text}"*
    *   **Use Case:** Extracting messy, unstructured data where the format constantly changes.
*   **Intent Classification Logic**
    *   **Core Prompt:** *"Analyze the user's input and classify their core intent into exactly ONE of the following categories: [QUOTE_REQUEST, CLAIM_SUBMISSION, GENERAL_INQUIRY, COMPLAINT]. Do not explain your reasoning. Output only the exact category name. Input: {text}"*
    *   **Use Case:** Figuring out what a user wants to do when they send an email or chat message.
*   **Sentiment / Tone Analysis Logic**
    *   **Core Prompt:** *"Evaluate the sentiment of the following customer message. Classify it as POSITIVE, NEUTRAL, or NEGATIVE. Also provide a frustration score from 1 to 10. Message: {text}"*
    *   **Use Case:** Prioritizing angry customer emails to the top of the customer service queue.
*   **Summarization / Synthesis Logic**
    *   **Core Prompt:** *"Read the following multi-page insurance policy document. Provide a 3-bullet point summary of the main coverage limits and exclusions. Document: {text}"*
    *   **Use Case:** Condensing a 20-page PDF into a quick summary for an agent to read.

### 🧩 AI Framework: The Plain-English Guide

Think of this framework like a **Smart Office** working to achieve a goal. It's not a simple robot that only understands 1s and 0s; it thinks, loops, and adapts.

*   **🤖 AGENT (The Boss / Persona)**
    *   **What it is:** The manager overseeing the operation.
    *   **Role:** It has a Goal (e.g., "Process this insurance quote"). It doesn't do the physical work itself, but it decides *who* and *what* is needed to get the job done.

*   **🗺️ WORKFLOW (The Roadmap / Process)**
    *   **What it is:** The path from Point A (Input) to Point B (Goal).
    *   **Role:** It tells the Agent the general steps to follow (e.g., Step 1: Read Document -> Step 2: Extract Data -> Step 3: Check for Errors).

*   **🧠 MEMORY (The 'Where are we?' / Context)**
    *   **What it is:** The office filing cabinet and the Agent's short-term memory.
    *   **Role:** It remembers past actions, feedback, and current status. It answers: *"What have we tried so far? Did it work?"* so the Agent doesn't repeat mistakes.

*   **🤔 LOGIC (The Reasoning / Brainpower)**
    *   **What it is:** The thinking and decision-making engine.
    *   **Role:** It analyzes the situation, looks at the Memory, and decides what to do next. It runs in a loop until it's confident it has the right answer.

*   **🗣️ PROMPT (The Instructions / Communication)**
    *   **What it is:** The detailed instruction manual or the language used to talk to the AI.
    *   **Role:** It tells the AI exactly *how* to think about a specific problem (e.g., "Read this text and find the Date").

*   **🛠️ TOOL (The Hands / Action)**
    *   **What it is:** The physical execution.
    *   **Role:** If the Logic decides it needs to open a PDF, the Tool is the actual script that rips open the file. It's the calculator, the search engine, or the file reader.

*   **🦸‍♂️ SKILL (The Superpower)**
    *   **What it is:** A complete package combining **Logic (Brain)** + **Prompt (Instructions)** + **Tool (Hands)**.
    *   **Role:** An ability the Agent has. For example, a "Document Reading Skill" knows *how to think* about documents (Logic), *how to open them* (Tool), and *what to look for* (Prompt).

````python
class ValidationLogic(BaseLogic):
    """
    Core Logic (Deterministic/Rule-Based): Validates extracted data against strict rules.
    """
    def __init__(self, rules: dict):
        """
        Args:
            rules (dict): A dictionary of functions returning True/False for each field.
        """
        self.rules = rules

    def analyze(self, context: dict, **kwargs) -> dict:
        """
        Evaluates extracted dictionary data against the deterministic rules.
        """
        validation_results = {}
        is_valid = True

        for field, value in context.items():
            if field in self.rules:
                # Run the strict deterministic rule (function)
                passed = self.rules[field](value)
                validation_results[field] = "PASS" if passed else "FAIL"
                if not passed:
                    is_valid = False
            else:
                validation_results[field] = "NO_RULE"

        return {
            "is_overall_valid": is_valid,
            "field_breakdown": validation_results
        }
````

## **2. LOGIC**

````python
from abc import ABC, abstractmethod

class BaseLogic(ABC):
    """
    Core Logic (Base Class): Defines the foundational thinking process or algorithm.
    It encapsulates the rules for parsing, decision-making, or structuring approaches.
    """
    @abstractmethod
    def analyze(self, context: str, **kwargs):
        """
        Must be implemented by subclasses to perform specific logic operations.
        """
        pass
````

````python
import re

class RegexExtractionLogic(BaseLogic):
    """
    Core Logic (Implementation): Extracts specific data points from text using regular expressions.
    """
    def __init__(self, patterns: dict):
        """
        Args:
            patterns (dict): A dictionary where keys are field names and values are regex strings.
        """
        self.patterns = {k: re.compile(v) for k, v in patterns.items()}

    def analyze(self, context: str, **kwargs) -> dict:
        """
        Extracts data based on the initialized regex patterns.

        Args:
            context (str): The text document to extract data from.

        Returns:
            dict: A dictionary of extracted key-value pairs.
        """
        extracted_data = {}
        for key, pattern in self.patterns.items():
            match = pattern.search(context)
            if match:
                # Return the first captured group if it exists, otherwise the whole match
                extracted_data[key] = match.group(1).strip() if match.groups() else match.group(0).strip()
            else:
                extracted_data[key] = "Not Found"

        return extracted_data
````
